Задан объект в пространстве 2D. Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Уменьшить объект в два раза.
B=np.array(A)*0.5
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Повернуть на 130 градусов.
# Перевод градусов в радианы: [pi/180]
# Используем линейные отображения
import math 
G = [[math.cos(130*np.pi/180), -math.sin(130*np.pi/180), 0],
     [math.sin(130*np.pi/180), math.cos(130*np.pi/180), 0],
     [0,0,1]
    ]
G = np.array(G)
print(G)

# В матрицу трансформации G добавь третий элемент в вектор строку (0), и добавь третий вектор [0, 0, 1]
```

```python
H = A@G
# print(H)
x = H[:,0]
y = H[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
