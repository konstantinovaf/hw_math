### Задание 3:

Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$


```python
def f(x):
    return x ** 5 + 4*np.sin(2*x) + np.cos(3*x+3)
# scipy.misc.derivative(func, x0, dx=1.0, n=1, args=(), order=3)
```

```python
# link to find derivatives
# https://docs.scipy.org/doc/scipy-0.18.0/reference/generated/scipy.misc.derivative.html
from scipy.misc import derivative
# The point at which n-th derivative is found.
x0 = 1

# Order of the derivative. Default is 1.
n3=3
n4=4

f3x =[derivative(f, x0=x0, n=n3, order = n3+2)]
f4x = [derivative(f, x0=x0, n=n4, order = n4+1)]
print(f3x, f4x)  
```
